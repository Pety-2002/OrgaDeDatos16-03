```python
import pandas as pd
encuestas = pd.read_csv("encuesta.csv")
```

Para que sea mas facil de manipular, renombro las columnas que me interesan a labels de una unica palabra en minuscula

```python
encuestas = encuestas.rename(columns={
    "Curso":"materia",
    "Opinión General Sobre el Curso":"nota",
    "¿Aprobó la Cursada?":"aprobo",
    "Nivel de las Clases Teóricas del Curso":"teoricas",
    "Nivel de las Clases Prácticas del Curso":"practicas",
    "¿Los Temas de la Materia Están Actualizados?":"actualizada"
})
encuestas
```

Convierto las columnas de cadena a entero (aquellas que tienen un valor dentro de una escala) y a `bool` las que son Si/No. 

```python
import numpy as np

traduccion = {
        'Excelente':5,
        'Excelentes':5,
        'Muy Bueno':4,
        'Muy Buenas':4,
        'Bueno':3,
        'Buenas':3,
        'Regular':2,
        'Regulares':2,
        'Malo':1,
        'Malas':1
    }


def mapear_a_entero(cadena):

    return traduccion.get(cadena,np.nan)

encuestas.nota = encuestas.nota.map(mapear_a_entero)
encuestas.teoricas = encuestas.teoricas.map(mapear_a_entero)
encuestas.practicas = encuestas.practicas.map(mapear_a_entero)
```

Para las columnas de Si/No, puedo utilizar directamente el operador `==` con la serie. Al utilizar un operador con una serie y un 'escalar' (un unico valor, no una serie), Pandas aplica la operacion elemento a elemento. Es similar a un `map` solo que funciona con operadores especificos. `map` por otro lado, permite aplicar funciones arbitrarias

```python
encuestas.aprobo = encuestas.aprobo == "Sí"
encuestas.actualizada = encuestas.actualizada == "Sí"
```

Extraemos el codigo de materia y nos quedamos con solo las columnas que nos importan

```python
encuestas.materia = encuestas.materia.map(lambda x:x.split("-")[0])
encuestas_sliced = encuestas[["materia","nota","aprobo","actualizada","teoricas","practicas"]]
encuestas_sliced
```

Queremos analizar como son las notas que se asignan a las materias, viendo si dependen de haber aprobado la materia o no. Para eso vamos generarnos 3 DataFrames con los datos nota promedio y cantidad de respuestas, uno con todas las encuestas, otra con solo las de alumnos aprobados y finalmente otra con solo las de alumnos desaprobados

```python
promedio_encuestas = encuestas.groupby("materia").agg({"nota":["mean","count"]})
promedio_aprobados = encuestas[encuestas.aprobo].groupby("materia").agg({"nota":["mean","count"]})
promedio_desaprobados = encuestas[~encuestas.aprobo].groupby("materia").agg({"nota":["mean","count"]})
```

Reseteamos el inidice para volver a llevar el codigo de la materia a una columna

```python
promedio_encuestas.reset_index(inplace=True)
promedio_aprobados.reset_index(inplace=True)
promedio_desaprobados.reset_index(inplace=True)
```

Renombramos las columnas para eliminar el multiindice

```python
promedio_encuestas.columns = ["materia","nota","cantidad"]
promedio_aprobados.columns = ["materia","nota","cantidad"]
promedio_desaprobados.columns = ["materia","nota","cantidad"]
```

Como voy a querer concatenar los DataFrames, necesito poder de alguna forma señalizar de donde provienen los datos (caso contrario, luego del `concat` no sabriamos a que corresponde cada promedio)

```python
promedio_encuestas["origen"] = "totales"
promedio_aprobados["origen"] = "aprobados"
promedio_desaprobados["origen"] = "desaprobados"
```

Ploteamos los datos e intentamos ver como se relacionan los datos entre si

```python
import plotly.express as px

df = pd.concat([promedio_encuestas,promedio_aprobados,promedio_desaprobados])

px.scatter(df,x="materia",y="nota",color="origen",size="cantidad")
```

Como en realidad verlo graficamente es complicado a medida que aumenta el volumen de datos, podemos utilizar `corr` para obtener la matriz de correlacion entre las variables

```python
encuestas_sliced.corr()
```

Podemos aprovechar la matriz de correlacion y ver que con `loc` puedo acceder a partes especificas del DataFrame, pero no solo para obtener datos, sino que para modificarlos tambien.

```python
correlacion = encuestas_sliced.corr()
correlacion.loc["nota","aprobo"] = 1
correlacion.loc["aprobo","nota"] = -100

correlacion
```

La asignacion de datos no tiene que seguir ninguna regla respecto del tipo de dato. No obstante, si se agrega un dato cuyo tipo no es compatible con el resto, Pandas va a cambiar el tipo de dato de la columna al tipo que mejor generalice todos los datos de la columna.

```python
correlacion.loc["aprobo","aprobo"] = "non"
correlacion
```

```python
correlacion.info()
```

Hay que saber, que cuando se cambia el tipo de la columna, cada elemento tambien cambia de forma implicita de tipo. Por esto es que si ahora la columna es de tipo object, al obtener la transpuesta, todos los elementos object de esa columna van a ocupar un lugar en todas las nuevas columnas. Por ende, todas las columnas van a pasar a ser de tipo object, porque no va a quedar alternativa para generalizar los datos.

```python
correlacion.T
```

```python
correlacion.T.info()
```

Del grafico observamos que la distribucion de cantidad de encuestas no es pareja entre los conjuntos de aprobados y desaprobados. 


Primero agrupamos por materia y si aprobaron, podemos contabilizar cuantas encuestas de que tipo hay para cada materia.

```python
tmp = encuestas_sliced.groupby(["materia","aprobo"]).count().reset_index()
tmp
```

A esto ahora le puedo aplicar pivot y asi obtener una representacion donde para cada materia tenga la cantidad de aprobados y desaprobados en la misma fila. Esta representacion nos deja operar facilmnente para por ejemplo, obtener el total de encuestas

```python
pivot = tmp.pivot("materia","aprobo","nota").fillna(0)
pivot.columns=["desaprobados","aprobados"]
pivot["total"] = pivot.aprobados + pivot.desaprobados
pivot
```

De todas formas, en vez de primero tener que pasar por un `groupby` y despues por un `pivot`, puedo utilizar directamente `pivot_table` que existe para estos casos

```python
pivot_table = encuestas_sliced.pivot_table(values="nota", index="materia", columns="aprobo", aggfunc='count', fill_value=0)
pivot_table.columns=["desaprobados","aprobados"]
pivot_table["total"] = pivot_table.aprobados + pivot_table.desaprobados
pivot_table
```

En esta representacion puedo incluso ordenar las materias por mayor cantidad de respuestas de aprobados

```python
pivot_table.sort_values(["aprobados","desaprobados","total"],ascending=False)
```

O calcular el porcentaje de respuestas que pertenecen a alumnos desaprobados

```python
pivot_table["porcentual"] = pivot_table.desaprobados / pivot_table.total
pivot_table.describe()
```

Esto nos da una idea de como se comportan las encuestas, porque podemos ver que al menos la mitad de las materias no cuentan con ninguna clase de respuesta de gente desaprobada. 

Sobre el DataFrame original, puedo ver tambien como se distribuyen las notas por cada materia. Podemos buscar aquellas encuestas que son outliers, es decir, son valores 'raros'. 

Si suponemos que todo el mundo pone casi la misma nota para la materia (porque consideramos que se puede calificar de una forma estandarizada y objetiva), todos deberian poner notas cercanas al promedio. Aquellos que esten lejos del promedio serian los 'outliers', que por ej podrian ser encuestas inventadas o gente poco sincera.

Necesitamos para eso tener para cada encuesta, cuanto es el promedio de la materia a la que esta contestando. Para este tipo de situaciones existe el transform, que nos genera los datos listos para integrar en nuestro DataFrame original.

```python
encuestas_sliced["promedio"] = encuestas_sliced.groupby("materia").transform("mean")["nota"]
encuestas_sliced["std_d"] = encuestas_sliced.groupby("materia").transform("std")["nota"]
encuestas_sliced
```

Notar que pandas se queja de que no uso `loc`. Si estoy trabajando sobre una copia de un DataFrame, los valores que cambie no van a actualizarse en el original (obviamente). El warning tiene dos objetivos

- Avisarnos que los datos no se van a modificar en el original en caso que busquemos hacer eso
- Avisarnos que si hicieramos estas cosas sobre el DataFrame original seria mas eficiente (fijense que seguimos teniendo en memoria el original, hacer una copia para agregarle cosas es claramente menos eficiente que agregarselas al original y operar directamente sobre ese)



Ahora con lo que obtuvimos de las transformaciones anteriores, puedo calcular operando con las columnas, si una encuesta es un outlier o no

```python
encuestas_sliced["outlier"] = (encuestas_sliced.nota - encuestas_sliced.promedio).abs() > (encuestas_sliced.std_d * 2)
```

```python
encuestas_sliced[encuestas_sliced.outlier]
```

Hay que tener siempre en mente que Pandas las cosas por defecto que nos da, por algo lo hace. El describe, nos da ademas del promedio la desviacion estandar, que bien si no aportase nada extra a los demas campos lo podria omitir y consumir menos tiempo. En este caso vemos en el ejemplo como la desviacion estandar nos ayuda a entender como se distribuyen los datos respecto del promedio, lo que nos da una idea de que tan `homogeneas` son las encuestas 

```python
df = pd.DataFrame([
    [5,5,5,5,1,1,1,1],
    [4,4,2,2,2,2,4,4],
    [3,3,3,3,3,3,3,3]
])

df.T.describe()
```

Podemos aprovechar el calculo anterior de outliers y ver con `%timeit` si es mas eficiente (en tiempo) operar sobre columnas directamente o utilizando`apply`

```python
def f (x):
    return abs(x.nota - x.promedio) <= (x.std_d * 2) 

%timeit encuestas_sliced.apply(f,axis=1)
```
